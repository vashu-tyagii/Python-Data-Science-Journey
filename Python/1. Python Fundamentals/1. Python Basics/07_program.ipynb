### **Conditional Statements in Python**

Conditional statements ka matlab hai code ko chalane ke liye shartein (conditions) rakhna. Agar koi condition `True` hoti hai, to ek code block chalta hai, aur agar `False` hoti hai, to doosra code block chalta hai. Isse hamare programs mein decision-making aati hai.

### **If Conditional Statement in Python**

Yeh sabse basic conditional statement hai. Agar `if` ke aage di gayi condition `True` hoti hai, to uske andar ka code chalta hai.

**Syntax:**

```python
if condition:
    # Code to be executed if the condition is True
```

**Example:**

```python
age = 20
if age >= 18:
    print("Aap vote de sakte hain.")
```

**Output:** `Aap vote de sakte hain.`

### **If-else Conditional Statements in Python**

Yeh `if` statement ka extended version hai. Agar `if` ki condition `True` nahi hoti hai, to `else` block ka code chalta hai.

**Syntax:**

```python
if condition:
    # Code for True condition
else:
    # Code for False condition
```

**Example:**

```python
age = 15
if age >= 18:
    print("Aap vote de sakte hain.")
else:
    print("Aap abhi vote nahi de sakte.")
```

**Output:** `Aap abhi vote nahi de sakte.`

### **elif Statement**

`elif` ka matlab hai "else if". Iska istemal tab hota hai jab aapko ek se zyada conditions check karni hon. `elif` ki condition tabhi check hoti hai jab uske upar ki saari conditions `False` ho chuki hon.

**Syntax:**

```python
if condition1:
    # Code...
elif condition2:
    # Code...
else:
    # Code...
```

**Example:**

```python
score = 85
if score >= 90:
    print("A grade")
elif score >= 80:
    print("B grade")
else:
    print("C grade")
```

**Output:** `B grade`

### **Nested if..else Conditional Statements in Python**

Jab ek `if-else` statement doosre `if-else` statement ke andar hota hai, to use nested `if-else` kehte hain. Yeh complex conditions ko handle karne ke liye use hota hai.

**Example:**

```python
num = 15
if num > 0:
    print("Number positive hai.")
    if num % 2 == 0:
        print("Aur yeh even number bhi hai.")
    else:
        print("Aur yeh odd number hai.")
else:
    print("Number positive nahi hai.")
```

**Output:** `Number positive hai.`
`Aur yeh odd number hai.`

### **Ternary Conditional Statement in Python**

Yeh ek single line ka shortcut hai `if-else` likhne ka. Isse code concise ho jaata hai.

**Syntax:**

```python
value_if_true if condition else value_if_false
```

**Example:**

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)
```

**Output:** `Adult`

### **Match-Case Statement in Python**

`match-case` statement (Python 3.10+) dusri languages ke `switch` statement jaisa hai. Iska use ek value ko alag-alag patterns se match karne ke liye hota hai.

**Syntax:**

```python
match subject:
    case pattern1:
        # Code...
    case pattern2:
        # Code...
    case _: # Default case
        # Code...
```

**Example:**

```python
day = "Monday"
match day:
    case "Monday":
        print("Aaj kaam ka pehla din hai.")
    case "Friday":
        print("Weekend shuru hone wala hai!")
    case _:
        print("Normal din hai.")
```

**Output:** `Aaj kaam ka pehla din hai.`
