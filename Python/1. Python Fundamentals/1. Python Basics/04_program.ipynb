### **Python Operators**

Operators woh special symbols hain jo values aur variables par operations perform karte hain. Jaise `+` ek operator hai jo do numbers ko jodta hai.

### **Types of Operators in Python**

Python mein alag-alag tarah ke operators hote hain:

  * Arithmetic Operators
  * Comparison Operators
  * Logical Operators
  * Bitwise Operators
  * Assignment Operators
  * Identity Operators
  * Membership Operators

### **Arithmetic Operators in Python**

Yeh operators mathematical calculations ke liye use hote hain.

| Operator | Naam | Kaam | Example |
| :--- | :--- | :--- | :--- |
| `+` | Addition | Do numbers ko jodta hai. | `5 + 3 = 8` |
| `-` | Subtraction | Ek number se doosre ko ghataata hai. | `5 - 3 = 2` |
| `*` | Multiplication | Do numbers ko multiply karta hai. | `5 * 3 = 15` |
| `/` | Division | Do numbers ko divide karta hai. | `5 / 2 = 2.5` |
| `%` | Modulus | Division ke baad bacha hua remainder deta hai. | `5 % 2 = 1` |
| `**`| Exponentiation| Power (ghaat) nikalta hai. | `5 ** 2 = 25` |
| `//` | Floor Division| Division ke baad poora number deta hai (decimal hata kar). | `5 // 2 = 2` |

### **Comparison of Python Operators**

Yeh operators do values ki tulna (compare) karte hain aur result mein `True` ya `False` dete hain.

| Operator | Kaam | Example |
| :--- | :--- | :--- |
| `==` | Barabar hai? | `5 == 5` (`True`) |
| `!=` | Barabar nahi hai? | `5 != 3` (`True`) |
| `>` | Bada hai? | `5 > 3` (`True`) |
| `<` | Chota hai? | `5 < 3` (`False`) |
| `>=` | Bada ya barabar hai? | `5 >= 5` (`True`) |
| `<=` | Chota ya barabar hai? | `5 <= 3` (`False`) |

### **Logical Operators in Python**

Yeh operators conditional statements ko combine karne ke liye use hote hain.

| Operator | Kaam | Example |
| :--- | :--- | :--- |
| `and` | Agar dono conditions `True` hain toh `True` | `(5 > 3) and (5 < 10)` (`True`) |
| `or` | Agar koi bhi ek condition `True` hai toh `True` | `(5 == 5) or (5 == 10)` (`True`) |
| `not` | Condition ko ulta kar deta hai | `not (5 > 3)` (`False`) |

### **Bitwise Operators in Python**

Yeh operators numbers par unke binary representation (0s aur 1s) ke level par kaam karte hain.

| Operator | Naam | Kaam | Example |
| :--- | :--- | :--- | :--- |
| `&` | AND | Har bit ko compare karta hai. | `5 & 3` (`1`) |
| `|` | OR | Agar koi bhi ek bit 1 hai toh result 1. | `5 | 3` (`7`) |
| `^` | XOR | Agar bits alag-alag hain toh result 1. | `5 ^ 3` (`6`) |
| `<<` | Left Shift| Bits ko left mein shift karta hai. | `5 << 1` (`10`) |

### **Assignment Operators in Python**

Yeh operators values ko variables mein assign karte hain. `+=` jaise operators shorthand hain.

| Operator | Example | Same as |
| :--- | :--- | :--- |
| `=` | `x = 5` | `x = 5` |
| `+=`| `x += 5` | `x = x + 5` |
| `-=`| `x -= 5` | `x = x - 5` |

### **Identity Operators in Python**

Yeh operators check karte hain ki kya do variables memory mein ek hi object ko refer kar rahe hain. `is` aur `==` mein farq hota hai.

**Example:**

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x
print(x is y)    # Output: False (values same hain, par memory alag hai)
print(x == y)    # Output: True (values barabar hain)
print(x is z)    # Output: True (dono ek hi object ko refer kar rahe hain)
```

### **Membership Operators in Python**

Yeh operators check karte hain ki kya koi value kisi sequence (jaise list ya string) mein maujood hai ya nahi.

| Operator | Kaam | Example |
| :--- | :--- | :--- |
| `in` | Value maujood hai? | `'h' in 'hello'` (`True`) |
| `not in`| Value maujood nahi hai?| `'z' not in 'hello'` (`True`) |

### **Ternary Operator in Python**

Yeh ek single line ka shortcut hai `if-else` statement likhne ka.

**Syntax:** `value_if_true if condition else value_if_false`

**Example:**

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)    # Output: Adult
```

### **Precedence and Associativity of Operators in Python**

  * **Precedence (Priority):** Jab ek hi expression mein kai operators hote hain, toh yeh batata hai ki kis operator ko pehle solve karna hai. Multiplication aur Division ki precedence, Addition aur Subtraction se zyada hoti hai.

      * **Example:** `5 + 2 * 3` ka result `11` hoga, na ki `21` (kyunki `*` pehle solve hota hai).

  * **Associativity (Direction):** Agar do operators ki precedence barabar hai, toh unhein kis direction se solve karna hai (left-to-right ya right-to-left).

      * **Example:** `10 / 2 * 5` ka result `25.0` hoga. Pehle `10 / 2` solve hoga (`5.0`), fir `5.0 * 5` solve hoga.
