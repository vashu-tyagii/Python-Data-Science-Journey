### **Input and Output in Python**

Programming mein **Input (I/O)** ka matlab hai program aur user ke beech communication.

  * **Output:** Program se user ko information dena. Iske liye `print()` function use hota hai.
  * **Input:** User se program ko information lena. Iske liye `input()` function use hota hai.

### **Taking input in Python**

User se data lene ke liye `input()` function ka use hota hai. Yeh function hamesha user input ko **string** (text) format mein return karta hai.

**Code:**

```python
name = input("Apna naam likhein: ")
print("Aapka naam hai:", name)
```

**Kaise kaam karta hai:**

1.  `input()` function screen par "Apna naam likhein: " message dikhayega.
2.  Program user se input ka wait karega.
3.  Jo bhi user likhega, woh `name` variable mein string format mein store ho jayega.

### **Printing Output using print() in Python**

`print()` function ka kaam screen par output dikhana hai. Iske andar hum text, variables, ya calculations ka result de sakte hain.

**Code:**

```python
print("Hello, Python!")
```

**Kaise kaam karta hai:**

  * `print()` ke andar diya gaya content seedha screen par print ho jaata hai.

### **Printing Variables**

`print()` function ka use karke hum variables ki value ko bhi display kar sakte hain. Aap variables ko text ke saath comma (`,`) se alag karke print kar sakte hain.

**Code:**

```python
name = "Rohit"
age = 25
print("Mera naam hai", name, "aur meri umar hai", age)
```

**Output:**
`Mera naam hai Rohit aur meri umar hai 25`

### **Take Multiple Input in Python**

Ek hi line mein multiple inputs lene ke liye `input()` ke saath `.split()` method ka use hota hai.

**Code:**

```python
name1, name2 = input("Do naam daalein, space se alag karke: ").split()
print("Pehla naam:", name1)
print("Dusra naam:", name2)
```

**Kaise kaam karta hai:**

  * `input()` ek single string leta hai (jaise "Alice Bob").
  * `.split()` is string ko space se alag karke `['Alice', 'Bob']` jaisi list bana deta hai.
  * `name1, name2` mein values automatically assign ho jaati hain.

### **How to Change the Type of Input in Python**

`input()` se liya gaya data hamesha string hota hai. Jab aapko numbers par math operations karni ho, toh aapko uska data type badalna padta hai. Is process ko **Type Casting** kehte hain.

### **Print Names in Python**

Naam ya text string hote hain, isliye `input()` se li hui value ko badalne ki zaroorat nahi padti.

**Code:**

```python
name = input("Naam likhein: ")
print("Naam:", name)
```

### **Print Numbers in Python**

Numbers lene ke liye `input()` ko `int()` function ke andar use karna padta hai.

**Code:**

```python
age = int(input("Umar likhein: "))
print("Umar:", age)
print(type(age)) # Output: <class 'int'>
```

### **Print Float/Decimal Number in Python**

Decimal numbers lene ke liye `input()` ko `float()` function ke andar use karte hain.

**Code:**

```python
price = float(input("Daam likhein: "))
print("Daam:", price)
print(type(price)) # Output: <class 'float'>
```

### **Output Formatting**

Output ko aasan aur saaf-suthra banane ke liye **f-strings** ka use karte hain.

**Code:**

```python
name = "Karan"
age = 28
print(f"Mera naam {name} hai aur meri umar {age} saal hai.")
```

**Output:**
`Mera naam Karan hai aur meri umar 28 saal hai.`
