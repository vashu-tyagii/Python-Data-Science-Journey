### **Python Keywords**

Python Keywords woh reserved words hote hain jinka programming language mein ek special matlab hota hai. Inka istemal Python ke code ke structure aur logic ko define karne ke liye kiya jaata hai. Aap in keywords ko variable ya function ke naam ke roop mein istemal nahi kar sakte.

### **Getting a List of all Python keywords**

Aap Python ke built-in `keyword` module ka use karke saare keywords ki list nikal sakte hain.

**Code:**

```python
import keyword

print(keyword.kwlist)
```

**Output:**

```
['False', 'None', 'True', 'and', 'as', 'assert', 'async', 'await', 'break', 'class', 'continue', 'def', 'del', 'elif', 'else', 'except', 'finally', 'for', 'from', 'global', 'if', 'import', 'in', 'is', 'lambda', 'nonlocal', 'not', 'or', 'pass', 'raise', 'return', 'try', 'while', 'with', 'yield']
```

### **How to Identify Python Keywords?**

  * **Syntax Highlighting:** Zyadatar code editors (jaise VS Code, Jupyter) mein keywords alag color mein highlight hote hain, isse unhe pehchanna aasan ho jaata hai.
  * **`keyword.iskeyword()`:** Aap `keyword` module ke `iskeyword()` function ka use karke bhi check kar sakte hain ki koi shabd keyword hai ya nahi.
    ```python
    import keyword
    print(keyword.iskeyword("if"))  # Output: True
    print(keyword.iskeyword("name"))# Output: False
    ```

### **What Happens if We Use Keywords as Variable Names?**

Agar aap keyword ko variable ka naam dene ki koshish karte hain, to Python **`SyntaxError`** dega, kyunki yeh ek galat syntax hai.

**Example:**

```python
# if = 10     # Yahan error aayega!
# print(if)
```

### **Categorization of Keywords**

Keywords ko unke kaam ke hisaab se categories mein baanta ja sakta hai:

  * **Value Keywords:** `True`, `False`, `None`
  * **Conditional Keywords:** `if`, `elif`, `else`
  * **Looping Keywords:** `for`, `while`, `break`, `continue`
  * **Function/Class Keywords:** `def`, `class`, `return`, `lambda`
  * **Import Keywords:** `import`, `from`, `as`
  * **Error Handling Keywords:** `try`, `except`, `finally`, `raise`
  * **Other Keywords:** `and`, `or`, `not`, `in`, `is`, `with`, `pass`

### **Difference Between Keywords and Identifiers**

| | **Keywords** | **Identifiers** |
| :--- | :--- | :--- |
| **Kaam** | Language ke structure ko define karte hain. | Variables, functions, classes ke naam hote hain. |
| **Nature**| Reserved aur fixed hote hain. | User-defined (programmer decide karta hai). |
| **Rules**| Inka istemal nahi kar sakte. | Naming rules (jaise number se shuru na ho). |

### **Difference Between Variables and Keywords**

| | **Variables** | **Keywords** |
| :--- | :--- | :--- |
| **Kaam** | Data store karte hain. | Code ke structure ko control karte hain. |
| **Nature**| Ek type ke Identifiers hote hain. | Reserved words hain, jinhein istemal nahi kar sakte. |
| **Example**| `name = "Rahul"` | `if`, `else`, `def` |
