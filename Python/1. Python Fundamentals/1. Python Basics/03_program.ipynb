### **Python Variables**

Variables ek tarah ke containers hote hain jo data ko store karte hain. Har variable ka ek naam hota hai aur usmein ek value hoti hai.

### **Rules for Naming Variables**

Variable ka naam rakhte waqt kuch zaroori rules ka palan karna chahiye:

  * Variable ka naam alphabet (`a-z`, `A-Z`) ya underscore (`_`) se shuru hona chahiye. Number se shuru nahi ho sakta.
  * Variable ke naam mein space nahi ho sakta.
  * Python **case-sensitive** hai, isliye `age` aur `Age` do alag-alag variables hain.
  * Aap Python ke reserved keywords (jaise `if`, `for`) ka istemal variable ke naam ke roop mein nahi kar sakte.

### **Assigning Values to Variables**

Variables ko value dene ke liye `=` sign ka istemal hota hai, jise **assignment operator** kehte hain.

### **Basic Assignment**

Yeh sabse seedha tareeka hai, jismein ek variable ko ek value di jaati hai.
**Code:**

```python
my_age = 30
first_name = "Rahul"
```

### **Dynamic Typing**

Python ek dynamically-typed language hai. Iska matlab hai ki aapko variable declare karte samay uska data type batane ki zaroorat nahi padti. Python value ke hisaab se type khud hi decide kar leta hai.

**Code:**

```python
my_var = 10           # my_var ab ek integer hai
print(type(my_var))   # Output: <class 'int'>

my_var = "Hello"      # ab my_var ek string hai
print(type(my_var))   # Output: <class 'str'>
```

### **Multiple Assignments**

Python mein aap ek hi line mein ek se zyada variables ko values assign kar sakte hain.

### **Assigning the Same Value**

Agar aapko kai variables ko ek hi value deni hai.
**Code:**

```python
x = y = z = "Python"
print(x)  # Output: Python
print(y)  # Output: Python
```

### **Assigning Different Values**

Agar aapko ek hi line mein alag-alag variables ko alag-alag values deni hain.
**Code:**

```python
name, age, city = "Priya", 28, "Mumbai"
print(name)  # Output: Priya
print(age)   # Output: 28
```

### **Type Casting a Variable**

Type casting ka matlab hai ek variable ke data type ko doosre data type mein badalna. Yeh zaroori hota hai jab aapko numbers par math operations karni ho jo `input()` se string format mein aate hain.

### **Basic Casting Functions**

  * **`int()`:** Kisi value ko integer mein badalta hai.
    **Code:**
    ```python
    num_str = "10"
    num_int = int(num_str)
    print(type(num_int)) # Output: <class 'int'>
    ```
  * **`float()`:** Kisi value ko decimal (float) mein badalta hai.
    **Code:**
    ```python
    num_int = 25
    num_float = float(num_int)
    print(num_float) # Output: 25.0
    ```
  * **`str()`:** Kisi value ko string mein badalta hai.
    **Code:**
    ```python
    num = 123
    num_str = str(num)
    print("Number is: " + num_str) # Output: Number is: 123
    ```

### **Getting the Type of Variable**

Kisi bhi variable ka data type pata karne ke liye `type()` function ka istemal hota hai.
**Code:**

```python
my_list = [1, 2, 3]
print(type(my_list)) # Output: <class 'list'>

my_dict = {"a": 1}
print(type(my_dict)) # Output: <class 'dict'>
```
