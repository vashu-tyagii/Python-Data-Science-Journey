### **Python Data Types**

Data types woh classification hai jo Python ko batati hai ki ek variable kis tarah ka data store kar raha hai. Yeh bahut zaroori hai kyunki data type ke hisaab se hi Python us data par alag-alag operations kar sakta hai.

Python mein kuch main data types hain: Numeric, Sequence, Mapping, Boolean, aur Set types. Binary types jaise `bytes` aur `bytearray` bhi hote hain.

### **1. Numeric Data Types in Python**

Yeh numbers se deal karte hain.

  * **`int` (Integer):** Poore numbers (without decimal).
    ```python
    age = 25
    ```
  * **`float` (Floating Point):** Decimal numbers.
    ```python
    pi = 3.14
    ```
  * **`complex` (Complex Number):** Real aur imaginary numbers ka combination.
    ```python
    complex_num = 3 + 4j
    ```

### **2. Sequence Data Types in Python**

Yeh ek sequence mein items ka collection store karte hain, jahan har item ka ek index (position) hota hai.

  * **String Data Type (`str`):** Text data, quotes mein hota hai. Strings **immutable** (change nahi ho sakte) hote hain.
    ```python
    name = "Python"
    print(name[0])  # Output: P
    ```
  * **List Data Type (`list`):** Ordered aur **changeable (mutable)** collection. Alag-alag data types ke items ho sakte hain. Square brackets `[]` mein hote hain.
    ```python
    my_list = [1, "hello", 3.14]
    my_list[0] = 100 # Change possible hai
    print(my_list) # Output: [100, 'hello', 3.14]
    ```
  * **Tuple Data Type (`tuple`):** Ordered, lekin **unchangeable (immutable)** collection. Parentheses `()` mein hote hain.
    ```python
    my_tuple = (10, "a", 20)
    # my_tuple[0] = 50 # Error aayega
    ```

### **3. Boolean Data Type in Python**

  * **`bool`:** Yeh sirf do values store karta hai: `True` ya `False`. Yeh mostly comparisons aur conditional logic mein use hota hai.
    ```python
    is_student = True
    print(10 > 5)   # Output: True
    ```

### **4. Set Data Type in Python**

  * **`set`:** Unordered collection jismein sirf unique (ek jaise nahi) items hote hain. Duplicates ko automatically hata deta hai. Curly braces `{}` mein hote hain. `frozenset` iska immutable version hai.
    ```python
    my_set = {1, 2, 3, 2}
    print(my_set) # Output: {1, 2, 3} (duplicate hat gaya)
    ```

### **5. Dictionary Data Type**

  * **`dict` (Mapping Type):** Unordered collection, jismein data **key: value** pairs mein store hota hai. Har key unique honi chahiye. Curly braces `{}` mein hote hain.
    ```python
    person = {"name": "Alice", "age": 30}
    print(person["name"]) # Output: Alice
    ```
