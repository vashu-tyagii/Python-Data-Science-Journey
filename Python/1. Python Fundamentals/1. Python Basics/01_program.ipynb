### **1. Python Introduction**

Python ek high-level, interpreted programming language hai. Iska matlab hai ki iski syntax (likhne ka tareeka) bahut aasan aur padhne mein English se milti-julti hai. Iska istemal web development, data science, machine learning, aur automation jaise alag-alag kshetron (fields) mein hota hai. Python ko sikhna beginners ke liye bahut aasan hai, isliye yeh kaafi popular hai.

### **2. Key Features of Python**

  * **Easy to Learn and Read:** Iski syntax simple hai, jisse code jaldi likha aur aasaani se samjha ja sakta hai.
  * **Interpreted Language:** Python code ko line-by-line execute karta hai, jisse debugging (galati dhoondhna) aasan ho jaata hai.
  * **Cross-platform:** Yeh Windows, macOS, Linux, aur dusre operating systems par bina kisi badlav ke chal sakta hai.
  * **Dynamically Typed:** Aapko variable banate samay uska data type batane ki zaroorat nahi hoti. Python khud hi samajh jaata hai ki variable mein kya store hai.
  * **Rich Standard Library:** Python mein pehle se hi bahut saare built-in modules aur packages aate hain, jisse aapko kai kaam ke liye code likhne ki zaroorat nahi padti.

### **3. Understanding "Hello World" Program in Python**

Yeh kisi bhi programming language ka sabse pehla program hota hai. Iska maqsad hai screen par "Hello, World\!" print karna.

**Code:**

```python
print("Hello, World!")
```

**Kaise kaam karta hai:**

  * `print()`: Yeh Python ka ek built-in function hai, jiska kaam screen par output dikhana hai.
  * `()`: Parentheses ke andar jo bhi content hota hai, `print()` function use display karta hai.
  * `"Hello, World!"`: Yeh ek **string** (text) hai, jise humne `print()` function ko display karne ke liye diya hai.

### **4. Indentation in Python**

Python mein, **indentation** ka matlab hai code block ko define karne ke liye ek line ki shuruat mein spaces ya tabs dena. Dusri languages ki tarah, yahan curly braces `{}` ka istemal nahi hota. Indentation ka sahi istemal bahut zaroori hai.

**Rules:**

  * Ek hi code block ke andar ki sabhi lines par ek hi tarah ki indentation honi chahiye.
  * Aam taur par, 4 spaces ka use indentation ke liye kiya jaata hai.

**Example (Sahi Indentation):**

```python
if 5 > 2:
    print("5 is greater than 2")
    print("This is inside the if block.")

# Output:
# 5 is greater than 2
# This is inside the if block.
```

**Example (Galat Indentation):**

```python
if 5 > 2:
    print("5 is greater than 2")
      print("This is wrong indentation.")  # Yahan error aayega
```

Galat indentation se **`IndentationError`** aata hai.
