### **Python For Loops**

`for` loop ka istemal kisi sequence (jaise list, string, tuple, etc.) ke har item par ek-ek karke iterate karne ke liye hota hai. Isse aap ek code block ko har item ke liye chala sakte hain.

### **Python For Loop Syntax**

`for` loop ka basic syntax yeh hai:

```python
for item in sequence:
    # Code to be executed for each item
```

  * **`item`:** Har iteration mein sequence ka ek-ek item is variable mein aata hai.
  * **`sequence`:** Woh list, string, ya koi aur iterable object, jis par loop chalana hai.

### **Python For Loop with String**

Aap `for` loop se ek string ke har character ko iterate kar sakte hain.

**Example:**

```python
for char in "Python":
    print(char)
```

**Output:**

```
P
y
t
h
o
n
```

### **Using range() with For Loop**

`range()` function ka use ek sequence of numbers banane ke liye hota hai, jise aap `for` loop mein istemal kar sakte hain.

  * `range(5)` -\> `0, 1, 2, 3, 4`
  * `range(2, 5)` -\> `2, 3, 4`
  * `range(0, 10, 2)` -\> `0, 2, 4, 6, 8`

**Example:**

```python
for i in range(3):
    print("Hello")
```

**Output:**

```
Hello
Hello
Hello
```

### **Control Statements with For Loop**

Control statements loop ke flow ko control karte hain. Ismein `continue`, `break` aur `pass` shamil hain.

### **Continue with For Loop**

`continue` statement jab chalta hai, to woh current iteration ko rok kar agle iteration par chala jaata hai.

**Example:**

```python
for i in range(5):
    if i == 2:
        continue  # Jab i=2 hoga, to yeh iteration skip ho jayegi
    print(i)
```

**Output:**

```
0
1
3
4
```

### **Break with For Loop**

`break` statement jab chalta hai, to woh loop ko turant rok deta hai.

**Example:**

```python
for i in range(5):
    if i == 3:
        break  # Jab i=3 hoga, to loop band ho jayega
    print(i)
```

**Output:**

```
0
1
2
```

### **Pass Statement with For Loop**

`pass` statement ka matlab hai "kuch mat karo". Iska istemal ek empty code block ke liye placeholder ke roop mein hota hai, jab syntax ke hisaab se block zaroori ho, par code abhi likha na ho.

**Example:**

```python
for i in range(5):
    # Abhi yahan kya karna hai, pata nahi hai
    pass
```

### **Else Statement with For Loops**

`for` loop ke saath `else` block ka use kiya ja sakta hai. `else` block tab chalta hai jab loop poora ho jaata hai, bina `break` statement ke.

**Example:**

```python
for i in range(3):
    print(i)
else:
    print("Loop poora ho gaya!")
```

**Output:**

```
0
1
2
Loop poora ho gaya!
```

### **Using Enumerate with for loop**

`enumerate()` function ka use loop mein index aur item dono ko ek saath paane ke liye hota hai. Yeh ek tuple `(index, item)` wapas deta hai.

**Example:**

```python
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")
```

**Output:**

```
Index 0: apple
Index 1: banana
Index 2: cherry
```

### **Nested For Loops in Python**

Ek loop ke andar doosra loop hone ko nested loop kehte hain. Iska istemal multi-dimensional data (jaise 2D lists) ya combinations par iterate karne ke liye hota hai.

**Example:**

```python
for i in range(2):       # Outer loop
    for j in range(3):   # Inner loop
        print(f"({i}, {j})")
```

**Output:**

```
(0, 0)
(0, 1)
(0, 2)
(1, 0)
(1, 1)
(1, 2)
```
